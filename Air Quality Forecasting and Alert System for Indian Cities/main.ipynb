```python
df['last_update'] = pd.to_datetime(df['last_update'], format='mixed')
```

```python
df.duplicated().sum()
```

```python
df['hour'] = df['last_update'].dt.hour
df['day'] = df['last_update'].dt.day
df['month'] = df['last_update'].dt.month
df['weekday'] = df['last_update'].dt.weekday
df['is_weekdend'] = df['weekday'].isin([5,6]).astype(int)
```

```python
df.isnull().sum()
```

```python
df['pollutant_min'] = df['pollutant_min'].fillna(df['pollutant_min'].mean())
df['pollutant_max'] = df['pollutant_max'].fillna(df['pollutant_max'].mean())
df['pollutant_avg'] = df['pollutant_avg'].fillna(df['pollutant_avg'].mean())
```

```python
df['pollutant_id'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['pollutant_id'] = le.fit_transform(df['pollutant_id'])
```

```python
dict(zip(le.classes_ ,le.transform(le.classes_)))
```

```python
df = df.sort_values(['station', 'pollutant_id', 'last_update'])
```

```python
df['lag_1'] = df.groupby(['station','pollutant_id'])['pollutant_avg'].shift(1)
df['lag_2'] = df.groupby(['station','pollutant_id'])['pollutant_avg'].shift(2)
df['lag_3'] = df.groupby(['station','pollutant_id'])['pollutant_avg'].shift(3)
```

```python
le_state = LabelEncoder()
le_city = LabelEncoder()
df['state'] = le.fit_transform(df['state'])
df['city'] = le.fit_transform(df['city'])
```

```python
# x = df.drop(['pollutant_avg', 'last_update', 'country', 'station'], axis=1)
x = df.drop(['pollutant_avg', 'last_update', 'country', 'station' ,'lag_1','lag_2','lag_3'], axis=1)
y = df['pollutant_avg']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, r2_score

lr = LinearRegression()
rf = RandomForestRegressor(n_estimators=200, random_state=42)
gb = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1)
nn = MLPRegressor(hidden_layer_sizes=(64,32), max_iter=500, random_state=42)

models = {'Linear':lr, 'RandomForest':rf, 'GradientBoosting':gb, 'NeuralNet':nn}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"{name} | MAE: {mean_absolute_error(y_test, y_pred):.2f} | R2: {r2_score(y_test, y_pred):.2f}")
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x_train, y_train)
```

```python
y_pred_xgb = xgb.predict(x_test)

print("Ensemble MAE:", mean_absolute_error(y_test, y_pred_xgb))
print("Ensemble R2:", r2_score(y_test, y_pred_xgb))
```

```python
xgb.score(x_test, y_test)
```

```python
from sklearn.ensemble import VotingRegressor

ensemble = VotingRegressor(
    estimators=[('rf',rf),('gb',gb),('nn',nn), ('xgb', xgb)]
)
ensemble.fit(x_train, y_train)
```

```python
y_pred_ens = ensemble.predict(x_test)

print("Ensemble MAE:", mean_absolute_error(y_test, y_pred_ens))
print("Ensemble R2:", r2_score(y_test, y_pred_ens))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(y_test.values[:100], label='Actual')
plt.plot(y_pred_ens[:100], label='Predicted', linestyle='dashed')
plt.legend()
plt.title('Actual vs Predicted Air Quality Levels')
plt.show()
```
